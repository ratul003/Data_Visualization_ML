```python
# To extract values from a text file to present in a 2D relationship
X, Y= [], []
for line in open ("matplot.txt","r"): # Open in read only mode
    values= [float(s) for s in line.split()] # Separating the values
    X.append(values[0])
    Y.append(values[1])

plt.plot(X,Y)
plt.show()
```

```python
# Distribution Plots
def pdf(X, mu, sigma):
    a= 1./(sigma * np.sqrt(2. * np.pi))
    b= -1./(2. * sigma ** 2)
    return a * np.exp(b* (X - mu) ** 2)

X= np.linspace(-6,6,1000)

for i in range(1):
    samples= np.random.standard_normal(50)
    mu, sigma = np.mean(samples), np.std(samples)
    plt.plot(X, pdf(X,mu, sigma),color= '0.75')
    
    plt.plot(X, pdf(X,0.,1.), color= "k")
    plt.show()
    
X= np.linspace(-6,6,1024)
plt.plot(X, pdf(X,0.,1.), color= "k", linestyle= "solid", marker= 'o', markevery= 50)
plt.plot(X, pdf(X,0.,.5), color= "k", linestyle= "dashed")
plt.plot(X, pdf(X,0.,.25), color= "k", linestyle= "dashdot", marker= 'o', markevery= 75)

plt.show()

import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris Data
iris= sns.load_dataset("iris")
sns.swarmplot(x= 'species', y= 'petal_length', data= iris)
plt.show()
    
```
